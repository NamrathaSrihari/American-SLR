```python
import cv2
import numpy as np
import os
import string

if not os.path.exists("words"):
    os.makedirs("words")
    os.makedirs("words/train")
    os.makedirs("words/test")
    os.makedirs("words/train/Love")
    os.makedirs("words/train/Money")
    os.makedirs("words/train/Pray")
    os.makedirs("words/train/Church")
    os.makedirs("words/train/I love you")
    os.makedirs("words/train/I hate you")
    os.makedirs("words/train/No")
    os.makedirs("words/train/Yes")
    os.makedirs("words/train/Fine")
    os.makedirs("words/train/okay")
    os.makedirs("words/train/Help")
    os.makedirs("words/train/Hello")
    os.makedirs("words/train/Sorry")
    

    os.makedirs("words/test/Love")
    os.makedirs("words/test/Money")
    os.makedirs("words/test/Pray")
    os.makedirs("words/test/Church")
    os.makedirs("words/test/I love you")
    os.makedirs("words/test/I hate you")
    os.makedirs("words/test/No")
    os.makedirs("words/test/Yes")
    os.makedirs("words/test/Fine")
    os.makedirs("words/test/okay")
    os.makedirs("words/test/Help")
    os.makedirs("words/test/Hello")
    os.makedirs("words/test/Sorry")
    
    
if not os.path.exists("words1"):
    os.makedirs("words1")
    os.makedirs("words1/train")
    os.makedirs("words1/test")
    os.makedirs("words1/train/Love")
    os.makedirs("words1/train/Money")
    os.makedirs("words1/train/Pray")
    os.makedirs("words1/train/Church")
    os.makedirs("words1/train/I love you")
    os.makedirs("words1/train/I hate you")
    os.makedirs("words1/train/No")
    os.makedirs("words1/train/Yes")
    os.makedirs("words1/train/Fine")
    os.makedirs("words1/train/okay")
    os.makedirs("words1/train/Help")
    os.makedirs("words1/train/Hello")
    os.makedirs("words1/train/Sorry")
    

    os.makedirs("words1/test/Love")
    os.makedirs("words1/test/Money")
    os.makedirs("words1/test/Pray")
    os.makedirs("words1/test/Church")
    os.makedirs("words1/test/I love you")
    os.makedirs("words1/test/I hate you")
    os.makedirs("words1/test/No")
    os.makedirs("words1/test/Yes")
    os.makedirs("words1/test/Fine")
    os.makedirs("words1/test/okay")
    os.makedirs("words1/test/Help")
    os.makedirs("words1/test/Hello")
    os.makedirs("words1/test/Sorry")
    
# Train or test
mode = 'train'
directory = 'words/' + mode + '/'
directory1 = 'words1/' + mode + '/'

cap = cv2.VideoCapture(0)

while True:
    _, frame = cap.read()
    # Simulating mirror image
    frame = cv2.flip(frame, 1)

    # Getting count of existing images
    count = {'Church': len(os.listdir(directory + "/Church")),
             'Fine': len(os.listdir(directory + "/Fine")),
             'Hello': len(os.listdir(directory + "/Hello")),
             'Help': len(os.listdir(directory + "/Help")),
             'I hate you': len(os.listdir(directory + "/I hate you")),
             'I love you': len(os.listdir(directory + "/I love you")),
             'Love': len(os.listdir(directory + "/Love")),
             'Money': len(os.listdir(directory + "/Money")),
             'No': len(os.listdir(directory + "/No")),
             'okay': len(os.listdir(directory + "/okay")),
             'Pray': len(os.listdir(directory + "/Pray")),
             'Sorry': len(os.listdir(directory + "/Sorry")),
             'Yes': len(os.listdir(directory + "/Yes"))
             }

    # Printing the count in each set to the screen
    cv2.putText(frame, "MODE : " + mode, (10, 50), cv2.FONT_HERSHEY_PLAIN, 1, (0, 255, 255), 1)
    cv2.putText(frame, "IMAGE COUNT", (10, 60), cv2.FONT_HERSHEY_PLAIN, 1, (0, 255, 255), 1)
    cv2.putText(frame, "CHURCH : " + str(count['Church']), (10, 80), cv2.FONT_HERSHEY_PLAIN, 1, (0, 255, 255), 1)
    cv2.putText(frame, "FINE : " + str(count['Fine']), (10, 100), cv2.FONT_HERSHEY_PLAIN, 1, (0, 255, 255), 1)
    cv2.putText(frame, "HELLO : " + str(count['Hello']), (10, 120), cv2.FONT_HERSHEY_PLAIN, 1, (0, 255, 255), 1)
    cv2.putText(frame, "HELP : " + str(count['Help']), (10, 140), cv2.FONT_HERSHEY_PLAIN, 1, (0, 255, 255), 1)
    cv2.putText(frame, "I HATE YOU : " + str(count['I hate you']), (10, 160), cv2.FONT_HERSHEY_PLAIN, 1, (0, 255, 255), 1)
    cv2.putText(frame, "I LOVE YOU : " + str(count['I love you']), (10, 180), cv2.FONT_HERSHEY_PLAIN, 1, (0, 255, 255), 1)
    cv2.putText(frame, "LOVE : " + str(count['Love']), (10, 200), cv2.FONT_HERSHEY_PLAIN, 1, (0, 255, 255), 1)
    cv2.putText(frame, "MONEY : " + str(count['Money']), (10, 220), cv2.FONT_HERSHEY_PLAIN, 1, (0, 255, 255), 1)
    cv2.putText(frame, "NO : " + str(count['No']), (10, 240), cv2.FONT_HERSHEY_PLAIN, 1, (0, 255, 255), 1)
    cv2.putText(frame, "OKAY : " + str(count['okay']), (10, 260), cv2.FONT_HERSHEY_PLAIN, 1, (0, 255, 255), 1)

    cv2.putText(frame, "PRAY : " + str(count['Pray']), (120, 80), cv2.FONT_HERSHEY_PLAIN, 1, (0, 255, 255), 1)
    cv2.putText(frame, "SORRY : " + str(count['Sorry']), (120, 100), cv2.FONT_HERSHEY_PLAIN, 1, (0, 255, 255), 1)
    cv2.putText(frame, "YES : " + str(count['Yes']), (120, 120), cv2.FONT_HERSHEY_PLAIN, 1, (0, 255, 255), 1)
  
    
    # Coordinates of the ROI
    x1 = int(0.5 * frame.shape[1])
    y1 = 10
    x2 = frame.shape[1] - 10
    y2 = int(0.5 * frame.shape[1])
    # Drawing the ROI
    # The increment/decrement by 1 is to compensate for the bounding box
    cv2.rectangle(frame, (x1 - 1, y1 - 1), (x2 + 1, y2 + 1), (255, 0, 0), 1)
    # Extracting the ROI
    roi = frame[y1:y2, x1:x2]
    roi = cv2.resize(roi, (64, 64))

    cv2.imshow("Frame", frame)
    cv2.imshow("roi",roi)
    # _, mask = cv2.threshold(roi, 200, 255, cv2.THRESH_BINARY)
    # kernel = np.ones((1, 1), np.uint8)
    # roi = cv2.dilate(mask, kernel, iterations=1)
    # roi = cv2.erode(mask, kernel, iterations=1)
    # do the processing after capturing the image!
    roi = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(roi, (5, 5), 2)
    th3 = cv2.adaptiveThreshold(blur, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 11, 2)
    ret, roi1 = cv2.threshold(th3, 70, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    # _, roi = cv2.threshold(roi, 120, 255, cv2.THRESH_BINARY)
    # roi1 = cv2.resize(roi, (128, 128))
    cv2.imshow("ROI", roi1)

    interrupt = cv2.waitKey(10)
    if interrupt & 0xFF == 27:  # esc key
        break
    if interrupt & 0xFF == ord('0'):
        cv2.imwrite(directory + 'Church/' + str(count['Church']) + '.jpg', roi1)
        cv2.imwrite(directory1 + 'Church/' + str(count['Church']) + '.jpg', roi)
    if interrupt & 0xFF == ord('1'):
        cv2.imwrite(directory + 'Fine/' + str(count['Fine']) + '.jpg', roi1)
        cv2.imwrite(directory1 + 'Fine/' + str(count['Fine']) + '.jpg', roi)
    if interrupt & 0xFF == ord('2'):
        cv2.imwrite(directory + 'Hello/'  + str(count['Hello']) + '.jpg', roi1)
        cv2.imwrite(directory1 + 'Hello/' + str(count['Hello']) + '.jpg', roi)
    if interrupt & 0xFF == ord('3'):
        cv2.imwrite(directory + 'Help/' + str(count['Help']) + '.jpg', roi1)
        cv2.imwrite(directory1 + 'Help/' + str(count['Help']) + '.jpg', roi)
    if interrupt & 0xFF == ord('4'):
        cv2.imwrite(directory + 'I hate you/' + str(count['I hate you']) + '.jpg', roi1)
        cv2.imwrite(directory1 + 'I hate you/' + str(count['I hate you']) + '.jpg', roi)
    if interrupt & 0xFF == ord('5'):
        cv2.imwrite(directory + 'I love you/' + str(count['I love you']) + '.jpg', roi1)
        cv2.imwrite(directory1 + 'I love you/' + str(count['I love you']) + '.jpg', roi)
    if interrupt & 0xFF == ord('6'):
        cv2.imwrite(directory + 'Love/' + str(count['Love']) + '.jpg', roi1)
        cv2.imwrite(directory1 + 'Love/' + str(count['Love']) + '.jpg', roi)
    if interrupt & 0xFF == ord('7'):
        cv2.imwrite(directory + 'Money/' + str(count['Money']) + '.jpg', roi1)
        cv2.imwrite(directory1 + 'Money/' + str(count['Money']) + '.jpg', roi)
    if interrupt & 0xFF == ord('8'):
        cv2.imwrite(directory + 'No/' + str(count['No']) + '.jpg', roi1)
        cv2.imwrite(directory1 + 'No/' + str(count['No']) + '.jpg', roi)
    if interrupt & 0xFF == ord('9'):
        cv2.imwrite(directory + 'okay/' + str(count['okay']) + '.jpg', roi1)
        cv2.imwrite(directory1 + 'okay/' + str(count['okay']) + '.jpg', roi)

    if interrupt & 0xFF == ord('a'):
        cv2.imwrite(directory+'Pray/'+str(count['Pray'])+'.jpg', roi1)
        cv2.imwrite(directory1 + 'Pray/' + str(count['Pray']) + '.jpg', roi)
    if interrupt & 0xFF == ord('b'):
        cv2.imwrite(directory+'Sorry/'+str(count['Sorry'])+'.jpg', roi1)
        cv2.imwrite(directory1 + 'Sorry/' + str(count['Sorry']) + '.jpg', roi)
    if interrupt & 0xFF == ord('c'):
        cv2.imwrite(directory+'Yes/'+str(count['Yes'])+'.jpg', roi1)
        cv2.imwrite(directory1 + 'Yes/' + str(count['Yes']) + '.jpg', roi)


cap.release()
cv2.destroyAllWindows()
```
